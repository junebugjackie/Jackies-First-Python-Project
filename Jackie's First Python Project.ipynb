```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv(r'C:\Users\jggib\OneDrive\Documents\HabitPivotTable.csv')
# opening the file
df.fillna(0, inplace=True)
# filling null values with 0
df.columns = ['Habits', 'Jan', 'Feb', 'Mar', 'Apr', 'Total']
# renaming columns
df.drop([0,1,2,22], axis=0, inplace=True)
# dropping irrelevant data
df2 = df.replace("Sum of ", '', regex=True)
# getting rid of "Sum of [Habit]" and replacing with space so it just reads "[Habit]"
cols = ['Jan', 'Feb', 'Mar', 'Apr', 'Total']
df2[cols] = df[cols].apply(pd.to_numeric, errors='coerce', axis=1)
# changing the datatype to numeric
df2.info()
# dataframe info
print(df2)
# actual data frame
plt.pie(df2['Total'], labels = df2['Habits'])
# creating pie chart of habits
plt.show()
```
